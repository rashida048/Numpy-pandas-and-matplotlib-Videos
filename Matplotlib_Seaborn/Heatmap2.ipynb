```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("nhanes_2015_2016.csv")
df.head()
```

```python
df.columns
```

```python
df = df[['BPXSY1', 'BPXDI1', 'BPXSY2', 'BPXDI2', 'BMXWT', 'BMXHT', 'BMXBMI']]
df.head()
```

```python
cor = df.corr()
cor
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(data = cor, annot = True, cmap = 'mako', center = 0)
```

```python
plt.figure(figsize=(8, 8))
sns.heatmap(data = cor, annot = True, cmap = 'mako', center = 0, vmin = 0.1, vmax = 0.95, cbar = False)
```
